For Each Attribute: (all numeric-valued)
   1. Number of times pregnant
   2. Plasma glucose concentration a 2 hours in an oral glucose tolerance test
   3. Diastolic blood pressure (mm Hg)
   4. Triceps skin fold thickness (mm)
   5. 2-Hour serum insulin (mu U/ml)
   6. Body mass index (weight in kg/(height in m)^2)
   7. Diabetes pedigree function
   8. Age (years)
   9. Class variable (0 or 1)

```python
# first neural network with keras tutorial
from numpy import loadtxt
from keras.models import Sequential
from keras.layers import Dense
import pandas as pd
from sklearn.model_selection import train_test_split

# load the dataset
#dataset = loadtxt('C:/Users/muhzu/PycharmProjects/pima-indians-diabetes-data.csv', delimiter=',')
# split into input (X) and output (y) variables
headerList=['no. of time pregnant','oral glucose','diastolic blood pressure','triceps skin thickness',
                        '2-hours serum insulin','body mass index','diabetes predigree ftn','age','target']
df = pd.read_csv('C:/Users/muhzu/PycharmProjects/pima-indians-diabetes-data.csv',names=headerList)

df.head()
# X = dataset[:,0:8]
# y = dataset[:,8]

```

```python
x = df.drop("target",axis=1)
y = df["target"]
```

```python
x[:5]
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2,random_state=42)
```

```python
# define the keras model
model = Sequential()
model.add(Dense(12, input_dim=8, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python

# compile the keras model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python


# fit the keras model on the dataset
model.fit(X_train, y_train, epochs=20
          , batch_size=10)
```

```python

# evaluate the keras model
loss_of_model, accuracy = model.evaluate(X_test, y_test)
print(f'loss of the model {loss_of_model},Accuracy: %.2f' % (accuracy*100))
```

```python
# make probability predictions with the model
predictions = model.predict(X_test)
# round predictions
# rounded = [round(X_test[0]) for x in predictions]
# rounded[:10]
predictions[:5]
```
